```python
from cryptography.fernet import Fernet
#Fernet key: Used for encryption and decryption with 'cryptography' library
```

```python
#Generate and save fernet_key
##generate a new fernet key for symmetric encrption
key = Fernet.generate_key()
##open a file named'fernet_key.txt' in write-binary mode('wb')
with open('fernet_key.txt', 'wb') as key_file:
    ##write the generated key to the file
    key_file.write(key)
```

```python
#Encrypt api key
##crrate a Fernet instance with the generated key
fernet = Fernet(key)
##convert API key to bytes and encrypts
api_key = 'key-to-encrypt'
encrypted_api_key = fernet.encrypt(api_key.encode())

#Save api key
with open('encrpted_api_key.txt','wb') as enc_file:
    enc_file.write(encrypted_api_key)
```
